Graphical Data Visualization through Matplotlib.

```python
import matplotlib.pyplot as plt
```

First simple line chart

```python
x = [1,2,3,4,5,6]
y = [34,23,56,45,67,41]
z = [23,34,45,56,41,67]

plt.plot(x,y)
plt.show()
```

Now we will see how to add label

```python
plt.plot(x,y)
plt.title('Simple Line Chart')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
plt.plot(x,y,label='line 1')
plt.plot(x,z,label='line 2')

plt.title('Doubled Line Chart')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
plt.plot(x,y,label='line 1')
plt.plot(x,z,label='line 2')
plt.axis([0,7,0,100])

plt.title('Doubled Line Chart 2')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn')

plt.plot(x,y,label='line 1',color='red',marker='o',linestyle='dashed',linewidth='2',markersize='7',alpha=0.6)
plt.plot(x,z,label='line 2',color='blue',marker='s',linestyle='dotted',linewidth='2',markersize='7',alpha=0.6)
plt.axis([0,7,0,100])

plt.title('Customized Line Chart')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```

**Vertically Stacked Subplots**

```python
fig, (ax1,ax2) = plt.subplots(2,1)
plt.suptitle('Vertically Stacked Subplots Line Chart')
ax1.plot(x,y,label='line 1',color='orange',marker='o',linestyle='dashed',linewidth='2',markersize='7')
ax2.plot(x,z,label='line 2',color='green',marker='s',linestyle='dotted',linewidth='2',markersize='7',alpha=0.7)
plt.show()
```

**Horizontally Stacked Subplots**

```python
fig, (ax1,ax2) = plt.subplots(1,2)
plt.suptitle('Horizontally Stacked Subplots Line Chart')
ax1.plot(x,y,label='line 1',color='orange',marker='o',linestyle='dashed',linewidth='2',markersize='7')
ax2.plot(x,z,label='line 2',color='green',marker='s',linestyle='dotted',linewidth='2',markersize='7',alpha=0.7)
plt.show()
```

**Multiple Graph in one Figure**

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8,9]
a = [10,30,20,40,60,50,20,70,10]
b = [20,60,40,80,60,20,40,80,20]
c = [30,50,10,70,90,10,50,10,90]
d = [10,90,30,70,20,60,40,50,10]


fig = plt.figure()

plt.subplot(2,2,1)
plt.plot(x,a,label='line 1',color='tab:red',marker='o',linestyle='dashed',linewidth='2',markersize='7')

plt.subplot(2,2,2)
plt.plot(x,b,label='line 2',color='tab:blue',marker='s',linestyle='dotted',linewidth='2',markersize='7')

plt.subplot(2,2,3)
plt.plot(x,c,label='line 3',color='tab:green',marker='D',linestyle='dashdot',linewidth='2',markersize='7')

plt.subplot(2,2,4)
plt.plot(x,d,label='line 4',color='tab:orange',marker='*',linestyle='solid',linewidth='1',markersize='9')

plt.show()
```
